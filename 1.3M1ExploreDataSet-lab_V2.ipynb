Print the mean age of the survey participants.


```python
ages = []                                 # create an empty list

for text in df["Age"]:                    # go through every age
    parts = text.split("-")               # split "35-44" into ["35", "44 years old"]
    low = int(parts[0])                   # first number ‚Üí 35
    high = int(parts[1].split()[0])       # second number ‚Üí 44  { part 1 of this list ["35", "44 years old"] and 0 of this list [44, years, old]
    middle = (low + high) / 2             # midpoint ‚Üí (35 + 44) / 2
    ages.append(middle)                   # save it in the list
print(ages)
# 3. Find the average of all these midpoints
mean_age = sum(ages) / len(ages)

print("Mean age of survey participants:", mean_age)
```
