## Identify and Analyze Duplicates


### Task 1: Identify Duplicate Rows
1. Count the number of duplicate rows in the dataset.
3. Display the first few duplicate rows to understand their structure.


```python
duplicate_count = df.duplicated().sum()
print("Number of duplicate rows:", duplicate_count)
```

```python
duplicate_rows = df[df.duplicated()]
duplicate_rows.head()
```

### Task 2: Analyze Characteristics of Duplicates
1. Identify duplicate rows based on selected columns such as MainBranch, Employment, and RemoteWork. Analyse which columns frequently contain identical values within these duplicate rows.
2. Analyse the characteristics of rows that are duplicates based on a subset of columns, such as MainBranch, Employment, and RemoteWork. Determine which columns frequently have identical values across these rows.
   


```python
#keep=False keeps ALL rows that have duplicates, instead of keeping one as original.
```

```python
subset_cols = ["MainBranch", "Employment", "RemoteWork"]
duplicate_subset = df[df.duplicated(subset=subset_cols, keep=False)]
duplicate_subset.head()
```

```python
unique_counts = duplicate_subset.nunique()
print(unique_counts)
```

```python
frequently_identical = unique_counts[unique_counts == 1]
print("Columns with identical values across duplicates:")
print(frequently_identical)
```

### Task 3: Visualize Duplicates Distribution
1. Create visualizations to show the distribution of duplicates across different categories.
2. Use bar charts or pie charts to represent the distribution of duplicates by Country and Employment.


```python
pip install seaborn
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count duplicates per Country
country_counts = duplicate_rows['Country'].value_counts().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=country_counts.values, y=country_counts.index, palette="coolwarm")
plt.title("Distribution of Duplicate Rows by Country")
plt.xlabel("Count of Duplicates")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

```python
# Count duplicate rows by Employment
employment_counts = duplicate_rows['Employment'].value_counts()

import matplotlib.pyplot as plt

# Example: duplicate distribution by Employment
labels = employment_counts.index.tolist()
sizes = employment_counts.values
colors = plt.cm.tab20.colors  # a colormap for distinct colors

fig, ax = plt.subplots(figsize=(12, 12))

# Create pie chart WITHOUT labels on slices
wedges, texts, autotexts = ax.pie(
    sizes,
    labels=None,            # remove labels on slices
    colors=colors[:len(sizes)],
    autopct='%1.1f%%',      # show percentages on slices
    startangle=140
)

# Add legend on the side with category names
ax.legend(
    wedges,
    labels,
    title="Employment Type",
    loc="center left",
    bbox_to_anchor=(1, 0, 0.5, 1)  # puts legend to the right
)

# Ensure the pie chart is a circle
ax.axis('equal')

plt.title("Distribution of Duplicate Rows by Employment")
plt.tight_layout()
plt.show()

```

### Task 4: Strategic Removal of Duplicates
1. Decide which columns are critical for defining uniqueness in the dataset.
2. Remove duplicates based on a subset of columns if complete row duplication is not a good criterion.


```python
# Define subset of key columns
subset_cols = ["MainBranch", "Employment", "RemoteWork"]

# Remove duplicates based on these columns, keep first occurrence
df_unique = df.drop_duplicates(subset=subset_cols, keep='first')

# Check how many rows were removed
removed_count = len(df) - len(df_unique)
print("Number of duplicates removed based on subset:", removed_count)
```
