### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
df['Country'].isna().sum()   #missing entries
```

```python
df['Country'].unique()       #Helps detect spelling mistakes and unusual values.
```

```python
df[df['Country'].str.contains(r'^\s|\s$', na=False)]
```

```python
df['Country'].value_counts().tail(10)
```

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
df['Country'] = df['Country'].str.strip().str.title()  #Remove extra spaces & fix capitalization

country_mapping = {
    'Usa': 'United States',
    'United States Of America': 'United States',
    'Uk': 'United Kingdom',
    'England': 'United Kingdom',
    'Scotland': 'United Kingdom',
    'Russia': 'Russian Federation',
    'South Korea': 'Korea, Republic Of'
}

df['Country'] = df['Country'].replace(country_mapping)     # Map inconsistent country name
df['Country'] = df['Country'].str.split(',').str[0]     # Handle multiple-country entries
```

```python
df['EdLevel'].unique()            #Inspect unique values

edlevel_mapping = {
    'Bachelor’s degree (B.A., B.S., B.Eng., etc.)': 'Bachelor',
    'Master’s degree (M.A., M.S., M.Eng., MBA, etc.)': 'Master',
    'Doctoral degree (Ph.D., Ed.D., etc.)': 'Doctorate',
    'Some college/university study without earning a degree': 'Some College',
    'Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)': 'High School',
    'Primary/elementary school': 'Primary School',
    'Associate degree (A.A., A.S., etc.)': 'Associate',
    'Professional degree (JD, MD, Ph.D, Ed.D, etc.)': 'Professional'

}

df['EdLevel'] = df['EdLevel'].replace(edlevel_mapping)         #Map to simplified categories

```

```python
df['EdLevel'].value_counts()               #verify results
```

### 4. Encoding Categorical Variables


<h5>4.1 Encode the Employment column using one-hot encoding.</h5>


```python
employment_encoded = pd.get_dummies(df['Employment'], prefix='Employment')
df = pd.concat([df, employment_encoded], axis=1)
#The Employment column was encoded using one-hot encoding, converting each employment category into a separate binary feature. This allows categorical employment data to be used effectively in machine learning models.
```

```python
df[employment_encoded]
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
missing_counts = df.isna().sum()
missing_counts_sorted = missing_counts.sort_values(ascending=False)
missing_counts_sorted[missing_counts_sorted > 0]
missing_counts_sorted.head(10)
#Columns with the highest number of missing values were identified by counting null entries in each column and sorting them in descending order. This helps determine which variables may require imputation or removal.
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
mean_value = df['ConvertedCompYearly'].mean()
df['ConvertedCompYearly'].fillna(mean_value, inplace=True)
```

```python
df['ConvertedCompYearly'].isna().sum()
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
mode_value = df['RemoteWork'].mode()[0]
df['RemoteWork'].fillna(mode_value, inplace=True)
```

```python
df['RemoteWork'].isna().sum()
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
df['ConvertedCompYearly_MinMax'] = (
    (df['ConvertedCompYearly'] - df['ConvertedCompYearly'].min()) /
    (df['ConvertedCompYearly'].max() - df['ConvertedCompYearly'].min())
)
```

```python
df['ConvertedCompYearly_MinMax'].describe()
#Min–Max Scaling was applied to the ConvertedCompYearly column to normalize compensation values between 0 and 1. This ensures that salary data is on a consistent scale and suitable for comparative analysis and machine learning models.
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
(df['ConvertedCompYearly'] <= 0).sum()   #Check for zero or negative values
```

```python
import numpy as np

df['ConvertedCompYearly_Log'] = np.log1p(df['ConvertedCompYearly'])  #Apply log transformation (recommended: log1p )
```

```python
df[['ConvertedCompYearly', 'ConvertedCompYearly_Log']].describe()
```

```python
df['ConvertedCompYearly'].skew(), df['ConvertedCompYearly_Log'].skew()
```

```python
#A logarithmic transformation was applied to the ConvertedCompYearly column using log1p to reduce right-skewness in compensation data. This helps stabilize variance and improves suitability for statistical analysis and modeling.
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
df['YearsCodePro'] = df['YearsCodePro'].replace({
    'Less than 1 year': 0.5,
    'More than 50 years': 50
})

df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')
#Clean and convert YearsCodePro to numeric
```

```python
def categorize_experience(years):
    if years <= 2:
        return 'Junior'
    elif years <= 5:
        return 'Mid'
    elif years <= 10:
        return 'Senior'
    else:
        return 'Expert'

df['ExperienceLevel'] = df['YearsCodePro'].apply(categorize_experience)
#create Experience level column
```

```python
df[['YearsCodePro', 'ExperienceLevel']].head()
```

```python
df['ExperienceLevel'].value_counts()
```

```python
#A new column ExperienceLevel was created by categorizing professional coding experience (YearsCodePro) into Junior, Mid, Senior, and Expert levels. This transformation simplifies experience-based analysis and improves interpretability.
```
