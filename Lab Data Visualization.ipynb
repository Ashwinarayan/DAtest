**Read the CSV File into a Pandas DataFrame**

**Create a SQLite Database and Insert the Data**

Now, let's create a new SQLite database (`survey-data.sqlite`) and insert the data from the DataFrame into a table using the sqlite3 library:


```python
import sqlite3

# Create a connection to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Write the dataframe to the SQLite database
df.to_sql('main', conn, if_exists='replace', index=False)


# Close the connection
conn.close()
```

**Verify the Data in the SQLite Database**
Verify that the data has been correctly inserted into the SQLite database by running a simple query:


```python
# Reconnect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Run a simple query to check the data
QUERY = "SELECT * FROM main LIMIT 5"
df_check = pd.read_sql_query(QUERY, conn)

# Display the results
print(df_check)
```

## Hands-on Lab


### Visualizing the Distribution of Data

**Histograms**

Plot a histogram of CompTotal (Total Compensation).


```python
import matplotlib.pyplot as plt
import sqlite3
import pandas as pd

# Fetch values
conn = sqlite3.connect('survey-data.sqlite')
query = "SELECT CompTotal FROM main WHERE CompTotal IS NOT NULL"
df = pd.read_sql_query(query, conn)
conn.close()

# Filter reasonable salary range 
df_clean = df[(df['CompTotal'] > 0) & (df['CompTotal'] < 500000)]

# Plot histogram
plt.figure(figsize=(10,6))
plt.hist(df_clean['CompTotal'], bins=50, color='skyblue', edgecolor='black')
plt.title('Distribution of Total Compensation (Filtered)')
plt.xlabel('Total Compensation')
plt.ylabel('Frequency')
plt.show()
```

**Box Plots**

Plot a box plot of Age.


```python
# Connect to SQLite
conn = sqlite3.connect('survey-data.sqlite')

# Read Age column
df_age = pd.read_sql_query("SELECT Age FROM main", conn)

# Close connection
conn.close()

# Clean and convert Age to numeric
# Many Age values come like '25-34', '45-54' etc.
# We’ll extract the lower bound as a representative number
df_age['Age_clean'] = df_age['Age'].str.extract(r'(\d+)', expand=False).astype(float)

# Drop NaN
df_age = df_age.dropna(subset=['Age_clean'])

# Plot boxplot
plt.figure(figsize=(8,6))
plt.boxplot(df_age['Age_clean'])
plt.title('Box Plot of Age')
plt.ylabel('Age (approximate)')
plt.grid(True)
plt.show()
```

### Visualizing Relationships in Data

**Scatter Plots**

Create a scatter plot of Age and WorkExp.


```python
# Connect to SQLite
conn = sqlite3.connect('survey-data.sqlite')

# Load Age and WorkExp columns
df = pd.read_sql_query("SELECT Age, WorkExp FROM main", conn)

# Close connection
conn.close()

# Clean and convert Age to numeric
df['Age_clean'] = df['Age'].str.extract(r'(\d+)', expand=False).astype(float)

# Convert WorkExp to numeric (invalid text -> NaN)
df['WorkExp_clean'] = pd.to_numeric(df['WorkExp'], errors='coerce')

# Drop rows where either is NaN
df_clean = df.dropna(subset=['Age_clean', 'WorkExp_clean'])

# Scatter plot
plt.figure(figsize=(10,6))
plt.scatter(df_clean['Age_clean'], df_clean['WorkExp_clean'], alpha=0.4, color='purple')
plt.title('Scatter Plot of Age vs Work Experience')
plt.xlabel('Age (approximate)')
plt.ylabel('Work Experience (years)')
plt.grid(True)
plt.show()
```

**Bubble Plots**

Create a bubble plot of `TimeSearching` and `Frustration` using the Age column as the bubble size.


```python
import numpy as np

# Select relevant columns and drop missing values
plot_df = df[['TimeSearching', 'Frustration', 'Age']].dropna()

# Convert Age ranges to numeric (midpoints)
age_map = {
    'Under 18 years old': 15,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70
}

plot_df['AgeNum'] = plot_df['Age'].map(age_map)

# Encode categorical variables as numbers
plot_df['TimeSearchingCode'] = plot_df['TimeSearching'].astype('category').cat.codes
plot_df['FrustrationCode'] = plot_df['Frustration'].astype('category').cat.codes
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(
    plot_df['TimeSearchingCode'],
    plot_df['FrustrationCode'],
    s=plot_df['AgeNum'] * 5,   # bubble size
    alpha=0.5
)

plt.xlabel('Time Searching (encoded)')
plt.ylabel('Frustration (encoded)')
plt.title('Bubble Plot of Time Searching vs Frustration (Bubble Size = Age)')

plt.grid(True)
plt.show()
```

### Visualizing Composition of Data

**Pie Charts**

Create a pie chart of the top 5 databases(`DatabaseWantToWorkWith`) that respondents wish to learn next year.


```python
import matplotlib.pyplot as plt
from collections import Counter

# Drop missing values
db_series = df['DatabaseWantToWorkWith'].dropna()

# Split multiple responses and flatten into a list
all_databases = []
for entry in db_series:
    all_databases.extend(entry.split(';'))

# Count occurrences
db_counts = Counter(all_databases)

# Get top 5 databases
top_5 = db_counts.most_common(5)

labels = [db[0] for db in top_5]
sizes = [db[1] for db in top_5]

# Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=140
)

plt.title('Top 5 Databases Respondents Want to Learn Next Year')
plt.show()
```

**Stacked Charts** 

Create a stacked bar chart of median `TimeSearching` and `TimeAnswering` for the age group 30 to 35.


```python
import matplotlib.pyplot as plt
import pandas as pd

# Filter age group (30–35 approximated as 25–34 years old)
age_df = df[df['Age'] == '25-34 years old'][['TimeSearching', 'TimeAnswering']].dropna()

# Ordered mapping for time categories
time_map = {
    'Less than 15 minutes a day': 10,
    '15-30 minutes a day': 22.5,
    '30-60 minutes a day': 45,
    '1-2 hours a day': 90,
    'More than 2 hours a day': 150
}

# Convert categorical values to numeric
age_df['TimeSearchingNum'] = age_df['TimeSearching'].map(time_map)
age_df['TimeAnsweringNum'] = age_df['TimeAnswering'].map(time_map)

# Calculate medians
median_searching = age_df['TimeSearchingNum'].median()
median_answering = age_df['TimeAnsweringNum'].median()

# Create stacked bar chart
plt.figure(figsize=(6, 6))
plt.bar(
    ['Age 30–35'],
    median_searching,
    label='Time Searching'
)

plt.bar(
    ['Age 30–35'],
    median_answering,
    bottom=median_searching,
    label='Time Answering'
)

plt.ylabel('Median Time (minutes/day)')
plt.title('Median Time Searching vs Answering (Age 30–35)')
plt.legend()

plt.show()
```

### Visualizing Comparison of Data

**Line Chart**

Plot the median `CompTotal` for all ages from 45 to 60.


```python
import matplotlib.pyplot as plt
import pandas as pd

# Filter relevant age groups
age_groups = ['45-54 years old', '55-64 years old']
salary_df = df[df['Age'].isin(age_groups)][['Age', 'CompTotal']].dropna()

# Calculate median compensation per age group
median_comp = salary_df.groupby('Age')['CompTotal'].median()

# Order age groups correctly
median_comp = median_comp.reindex(age_groups)

# Create line chart
plt.figure(figsize=(8, 5))
plt.plot(
    median_comp.index,
    median_comp.values,
    marker='o'
)

plt.xlabel('Age Group')
plt.ylabel('Median Total Compensation')
plt.title('Median Total Compensation for Ages 45–60')
plt.grid(True)

plt.show()
```

**Bar Chart**

Create a horizontal bar chart using the `MainBranch` column.


```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Query to get the count of each MainBranch
query = """
SELECT MainBranch, COUNT(*) as Count
FROM main
GROUP BY MainBranch
ORDER BY Count DESC
"""
df = pd.read_sql_query(query, conn)

# Close the connection
conn.close()

# Create a horizontal bar chart
plt.figure(figsize=(10,6))
plt.barh(df['MainBranch'], df['Count'], color='skyblue')
plt.xlabel('Number of Respondents')
plt.ylabel('Main Branch')
plt.title('Number of Respondents by Main Branch')
plt.gca().invert_yaxis()  # Optional: largest bar at the top
plt.show()
```
